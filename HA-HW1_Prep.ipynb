```python
train = pd.merge(df_train, df_trans, on='PatientGuid',how = 'left')
train.shape
```

### Get Dummies

```python
def get_dummy(df, col):
    df = pd.concat([df, pd.get_dummies(df[col], drop_first=True)], axis=1)
    df.drop(columns=col,inplace = True)
    return df 
```

```python
train = get_dummy(train,['Gender'])
```

```python
train = train.groupby(['PatientGuid']).first()
train= train[cols]
train.shape
```

```python
train.isna().sum()
```

```python
train.describe()
```

```python
train.head()
```

### Transform Zeros to Null

```python
train[['Weight']] = train[['Weight']].replace(0, np.nan)
df_temp = train.drop(columns = 'BMI')
```

```python
df_temp.head()
```

### Impute Null Values

```python
def impute_null(df):
    from sklearn.impute import KNNImputer
    import numpy as np

    # Create an instance of the KNNImputer class
    imputer = KNNImputer(n_neighbors=5)

    # Fit the imputer to the dataset
    imputer.fit(df)

    # Use the imputer to transform the dataset and replace missing values with imputed values
    output_array = imputer.transform(df)
    return output_array
```

```python
output_array = impute_null(df_temp)
output_array
```

### Insert Impute Values into Dataframe

```python
train['Weight'] = output_array[:, 3]
train['RespiratoryRate']=output_array[:, -3]
train['Temperature'] = output_array[:, -2]
```

```python
train.head()
```

### Calculate BMI

```python
train['BMI'] = train['Weight'] / (train['Height'] ** 2) * 703
```

```python
train.head()
```

### Output Dataframe

```python
train.to_csv('train.csv',index=True)
```
